```python
import pandas as pd
import numpy as np 
import lyricsgenius
import time
```

```python
token = "QDMHDVlmjwHp0BgJIXe7_UZHPe0w8Td3-POjRsmLtBFix4sRhnZOQWnoSOUVeO2E"
```

```python
genius = lyricsgenius.Genius(token, skip_non_songs=True, excluded_terms=["(Remix)", "(Live)"], remove_section_headers=True)
```

```python
artists = ["Morgan Wallen", "Luke Combs", "Kane Brown", "Jason Aldean",
           "Chris Stapleton", "Luke Bryan", "Carrie Underwood", "Florida Georgia Line",
           "Blake Shelton", "Tyler Childers", "Kenny Chesney", "Miranda Lambert",
           "Tim McGraw", "Eric Church", "Koe Wetzel", "Cody Johnson", "Toby Keith",
           "Gabby Barrett", "Keith Urban", "Zac Brown Band", "Thomas Rhett",
           "Dan + Shay", "Sam Hunt", "Rascal Flatts", "Josh Turner", "Brantley Gilbert",
           "Little Big Town", "Lee Brice", "Chris Young", "Parker McCollum",
           "Brett Young", "Zach Bryan", "Cody Jinks", "Brad Paisley", "Whiskey Myers",
           "Jon Pardi", "Maren Morris", "Dixie Chicks", "Kelsea Ballerini", "Old Dominion",
           "Lady Antebellum", "HARDY", "Jordan Davis", "Niko Moon", "Riley Green",
           "Billy Currington", "Darius Rucker", "Trace Adkins", "Cole Swindell", "Justin Moore"]
```

```python
start = time.time()
lyrics = []
artist_name = []
title = []
genre = []

for artist in artists:
    while True:
        try:
            songs = genius.search_artist(artist, max_songs = 25, sort = 'popularity').songs
            break
        except:
            pass
    for song in songs:
        s = song.lyrics
        s = s.replace('\n', ' ')
        lyrics.append(s)
        artist_name.append(artist)
        genre.append('Country')
        title.append(song.title)

Country = pd.DataFrame()
Country['Song'] = title
Country['Artist'] = artist_name
Country['Genre'] = genre
Country['Lyrics'] = lyrics

Country.to_csv('./Country.csv')

end = time.time()
```
